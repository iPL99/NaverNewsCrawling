```python
import time
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

# ChromeDriver 설정
chrome_options = Options()
# chrome_options.add_argument("--headless")  # 브라우저 창을 띄우지 않으려면 주석 해제
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)
driver.maximize_window()

# 날짜 범위 설정
start_date = datetime.strptime('20240713', '%Y%m%d')
end_date = datetime.strptime('20240813', '%Y%m%d')
delta = timedelta(days=1)

data = []

def wait_for_element(css_selector, timeout=10):
    """특정 CSS 셀렉터를 가진 요소가 로드될 때까지 대기"""
    return WebDriverWait(driver, timeout).until(EC.presence_of_element_located((By.CSS_SELECTOR, css_selector)))

def wait_for_page_to_load():
    """페이지가 완전히 로드될 때까지 대기"""
    WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#ct > div.press_ranking_home > div:nth-child(3) > ul')))

# 날짜 범위 반복
current_date = start_date
while current_date <= end_date:
    formatted_date = current_date.strftime('%Y%m%d')
    news_url = f'https://media.naver.com/press/081/ranking?type=popular&date={formatted_date}'
    
    # 날짜별 랭킹 페이지 로드
    driver.get(news_url)
    wait_for_page_to_load()
    
    for rank in range(1, 11):  # 기사 순위 1부터 10까지
        attempt = 1
        while attempt:
            try:
                # 뉴스 기사 링크 찾기 및 클릭
                news_selector = f'#ct > div.press_ranking_home > div:nth-child(3) > ul > li:nth-child({rank}) > a'
                element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, news_selector)))
                driver.execute_script("arguments[0].click();", element)
                
                # 클릭한 기사 페이지 로딩 대기
                WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#title_area')))
                
                # 기사 정보 추출
                title_selector = '#title_area > span'
                title_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, title_selector)))
                
                reporter_selector = 'em.media_end_head_journalist_name'
                section_selector = 'em.media_end_categorize_item'
                
                # 기자 이름 추출 (존재하지 않을 경우 기본값 사용)
                try:
                    reporter_element = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.CSS_SELECTOR, reporter_selector)))
                    reporter_name = reporter_element.text
                except:
                    reporter_selector = 'span.byline_s'
                    reporter_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, reporter_selector)))
                    reporter_name = reporter_element.text
                
                # 기사 섹션 추출
                section_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, section_selector)))
                
                title = title_element.text               
                article_section = section_element.text
                
                data.append({
                    '언론사': '서울신문',  # 실제 언론사 이름으로 수정
                    '날짜': formatted_date,
                    '랭킹': rank,
                    '기사 제목': title,
                    '기자 이름': reporter_name,
                    '기사 섹션': article_section
                })
                
                # 돌아가기
                driver.back()
                
                # 원래 페이지로 돌아올 때까지 대기
                wait_for_page_to_load()
                break
            except Exception as e:
                attempt += 1
                print(f"Attempt {attempt} failed for article rank {rank} on {formatted_date}: {e}")

    # 날짜를 하루씩 증가
    current_date += delta
```

```python
import pandas as pd
import numpy as np

# CSV 파일 읽기
daily_articles_df = pd.read_csv('daily_top10_articles.csv', encoding='UTF-8')
reporter_info_df = pd.read_csv('reporter_info.csv', encoding='UTF-8')

# 월간 기사 작성 건수를 정수형으로 변환하는 함수
def clean_article_count(value):
    try:
        # 소숫점이 있을 경우 정수로 변환
        return int(float(value))
    except (ValueError, TypeError):
        return None

# 월간 기사 작성 건수 컬럼을 정수형으로 변환
reporter_info_df['월간 기사 작성 건수'] = reporter_info_df['월간 기사 작성 건수'].apply(clean_article_count)

# CSV 파일의 기자 이름 컬럼명을 확인하고 동일하게 설정합니다. 예를 들어, '기자_이름'이라는 컬럼명이 있다고 가정합니다.
# 만약 컬럼명이 다르다면, 해당 컬럼명으로 변경하세요.

# 두 DataFrame을 '기자_이름' 컬럼을 기준으로 병합합니다.
merged_df = pd.merge(daily_articles_df, reporter_info_df, left_on='기자 이름', right_on='기자 이름', how='left')

# 결과를 새로운 CSV 파일로 저장합니다.
merged_df.to_csv('merged_articles_with_reporter_info.csv', index=False, encoding='utf-8-sig', na_rep='None')

print("Merged data has been saved to 'merged_articles_with_reporter_info.csv'")
```
