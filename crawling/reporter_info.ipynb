```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time

# ChromeDriver 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
url = "https://media.naver.com/journalists/whole?officeId=081"

# 페이지 열기
driver.get(url)

# 대기 시간
time.sleep(5)  # 초기 페이지 로딩 대기

# 링크 요소를 저장할 집합 (중복 방지)
links = set()

# 페이지 스크롤 반복하여 모든 기자 링크 가져오기
last_height = driver.execute_script("return document.body.scrollHeight")

while True:
    # 링크 요소 찾기
    new_links = driver.find_elements(By.CSS_SELECTOR, "#ct > div > div > section > div > div > div > ul > li > div > a.journalist_list_content_thumb")
    
    # 새로 찾은 링크 추가
    new_links_href = [link.get_attribute("href") for link in new_links]
    new_links_set = set(new_links_href)
    
    # 새로운 링크가 더 이상 없으면 종료
    if new_links_set.issubset(links):
        break
    
    # 기존 링크에 새 링크 추가
    links.update(new_links_set)
    
    # 스크롤을 아래로 내리기
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
    # 새로운 페이지의 높이 로드 확인
    time.sleep(10)  # 로딩 대기
    new_height = driver.execute_script("return document.body.scrollHeight")
    
    # 페이지가 더 이상 스크롤되지 않으면 종료
    if new_height == last_height:
        break
    
    last_height = new_height

# 링크 출력
for link in links:
    print("주소: ", link)
    print()

print("가져온 링크 수: ", len(links))

# 브라우저 종료
driver.quit()

# 링크를 파일에 저장
with open('reporter_links.txt', 'w', encoding='utf-8') as f:
    for link in links:
        f.write(link + '\n')

print("Reporter links have been saved to 'reporter_links.txt'")
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np  # NaN 처리를 위한 라이브러리

# ChromeDriver 설정
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# 기자 링크 파일 읽기
with open('reporter_links.txt', 'r', encoding='utf-8') as file:
    reporter_links = file.read().splitlines()

# 기자 정보 저장할 리스트
reporter_data = []

def extract_reporter_details(url):
    """기자 페이지에서 이름과 월간 기사 작성 건수 추출"""
    driver.get(url)
    time.sleep(0.1)  # 페이지 로딩 대기
    
    try:
        # BeautifulSoup을 사용하여 페이지 파싱
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        # 기자 이름 셀렉터
        name_selector = '#ct > div > div > section.journalist_sidebar > div > div.journalist_profile > div.journalist_profile_body > div > div > div.media_reporter_profile_inner > div.media_reporter_profile_basic > div.media_reporter_basic_text > h2'
        
        # 기자 이름 추출
        name_element = soup.select_one(name_selector)
        reporter_name = name_element.text.strip() + " 기자" if name_element else "정보 없음 기자"
        
        # 월간 기사 작성 건수 셀렉터
        count_selector = 'ul.media_reporter_summary_list > li.media_reporter_summary_item > em'
        
        # 월간 기사 작성 건수 추출
        li_tag = soup.find('li', class_='media_reporter_summary_item')
        if li_tag:
            monthly_article_count = li_tag.find('em').text.replace('총 ', '').strip()
        else:
            monthly_article_count = np.nan
        
        return reporter_name, monthly_article_count

    except Exception as e:
        print(f"Failed to extract details for {url}: {e}")
        return "정보 없음 기자", np.nan

# 기자 링크 반복 처리
for link in reporter_links:
    reporter_name, article_count = extract_reporter_details(link)
    
    reporter_data.append({
        '기자 이름': reporter_name,
        '월간 기사 작성 건수': article_count
    })

# 브라우저 종료
driver.quit()
```
